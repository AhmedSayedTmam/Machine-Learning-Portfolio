## Cleaning Data

#### Twitter archive Cleaning

```python
twitter_archive_enhanced_clean = twitter_archive_enhanced.copy()
```

#### Define
- Remove all rows that have values (not blank or non-null) in retweeted_status_id, retweeted_status_user_id, and         retweeted_status_timestamp

```python
twitter_archive_enhanced_clean = twitter_archive_enhanced_clean[twitter_archive_enhanced_clean.retweeted_status_id.isnull()]
twitter_archive_enhanced_clean = twitter_archive_enhanced_clean[twitter_archive_enhanced_clean.retweeted_status_user_id.isnull()]
twitter_archive_enhanced_clean = twitter_archive_enhanced_clean[twitter_archive_enhanced_clean.retweeted_status_timestamp.isnull()]
```

```python
twitter_archive_enhanced_clean = twitter_archive_enhanced_clean.drop('retweeted_status_id', axis=1)
twitter_archive_enhanced_clean = twitter_archive_enhanced_clean.drop('retweeted_status_user_id', axis=1)
twitter_archive_enhanced_clean = twitter_archive_enhanced_clean.drop('retweeted_status_timestamp', axis=1)
```

#### Test

```python
twitter_archive_enhanced_clean.info()
```

#### Define
- Separate between time and date

```python
# converting to datetime
twitter_archive_enhanced_clean['timestamp'] = pd.to_datetime(twitter_archive_enhanced_clean['timestamp'])

twitter_archive_enhanced_clean['time'] = twitter_archive_enhanced_clean['timestamp'].dt.time
twitter_archive_enhanced_clean['dates'] = twitter_archive_enhanced_clean['timestamp'].dt.date
```

```python
# dropping timestamp
twitter_archive_enhanced_clean = twitter_archive_enhanced_clean.drop('timestamp', axis=1)
```

#### Test

```python
twitter_archive_enhanced_clean.info()
```

#### Define
- Convert datatype for in_reply_to_status_id and in_reply_to_user_id to int     

```python
twitter_archive_enhanced_clean['in_reply_to_status_id'] = twitter_archive_enhanced_clean['in_reply_to_status_id'].fillna(0)
twitter_archive_enhanced_clean['in_reply_to_user_id'] = twitter_archive_enhanced_clean['in_reply_to_user_id'].fillna(0)
twitter_archive_enhanced_clean['in_reply_to_status_id'] = twitter_archive_enhanced_clean['in_reply_to_status_id'].astype(int)
twitter_archive_enhanced_clean['in_reply_to_user_id'] = twitter_archive_enhanced_clean['in_reply_to_user_id'].astype(int)
```

#### Test

```python
twitter_archive_enhanced_clean.info()
```

#### Define
- Convert datatype for rating_numerator and rating_denominator into float           

```python
twitter_archive_enhanced_clean['rating_numerator'] = twitter_archive_enhanced_clean['rating_numerator'].astype(float)
twitter_archive_enhanced_clean['rating_denominator'] = twitter_archive_enhanced_clean['rating_denominator'].astype(float)
```

#### Test

```python
twitter_archive_enhanced_clean.info()
```

#### Define
- Correct rating_denominator for values not equal to 10          

```python
# iterating through all rating_denominator is not 10
for i, row in twitter_archive_enhanced_clean[twitter_archive_enhanced_clean.rating_denominator != 10].iterrows():
    denominator = row.rating_denominator
    if denominator > 10 and denominator % 10 == 0:
        divisor = denominator/10
        numerator = row.rating_numerator
        # if rating_numerator is greater than 10 and divisible by the divisor
        if numerator % divisor == 0:
            # set rating_denominator as 10
            twitter_archive_enhanced_clean.at[i, 'rating_denominator'] = 10
            # set rating_numerator as numerator/divisor
            twitter_archive_enhanced_clean.at[i, 'rating_numerator'] = numerator/divisor
```

#### Test

```python
twitter_archive_enhanced_clean.query('rating_denominator != 10')
```

```python
twitter_archive_enhanced_clean.rating_denominator.value_counts()
```

#### Define
- Replace dogs name starting with lowercase characters like a, an with NaN

```python
twitter_archive_enhanced_clean['name'][twitter_archive_enhanced_clean['name'].str.match('[a-z]+')] = 'None'
```

#### Test

```python
twitter_archive_enhanced_clean.name.value_counts().sort_index(ascending=False)
```

#### Test

```python
twitter_archive_enhanced_clean.info()
```

#### Define
- Merge doggo, floofer, pupper and puppo columns into one column named "dog_stage"

```python
twitter_archive_enhanced_clean = pd.melt(twitter_archive_enhanced_clean,id_vars=['tweet_id', 'in_reply_to_status_id','in_reply_to_user_id',
                                                                                 'source', 'text', 'expanded_urls', 'rating_numerator',
                                                                                 'rating_denominator', 'name', 'time', 'dates'],
                                                                                 var_name='dogs', value_name='dog_stage')
```

```python
#twitter_archive_enhanced_clean = twitter_archive_enhanced_clean.drop('dogs', axis= 1)
# Dropping duplicated based on tweet_id
twitter_archive_enhanced_clean = twitter_archive_enhanced_clean.sort_values('dog_stage').drop_duplicates(subset='tweet_id', 
                                                                                        keep='last')
```
